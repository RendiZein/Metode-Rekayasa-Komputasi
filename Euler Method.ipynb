```python
#defined variable
def euler_no_1(delta_t=0.1):
    f_t_y=lambda y : 2-y
    y_0=5
    t_0=0
    t_end=1
    y_true_func=lambda t :2+3*np.exp(-t)
    y_true=[y_0]
    y_pred=[y_0]
    H = int((t_end-t_0)/delta_t)
    t=np.linspace(t_0, t_end, H+1)
    #looping
    for t_i in t[1:]:
        #calculate y_true
        y_true_i=y_true_func(t_i)
        y_true.append(y_true_i)
        #calculate y_pred
        y_pred_i=y_pred[-1]+f_t_y(y_pred[-1])*delta_t
        y_pred.append(y_pred_i)
    print('y_true value is \n{:0.5f} \ny_pred is\n{:0.5f}'.format(y_true[-1], y_pred[-1]))
    print('the error is {0:0.5f}%'.format(np.abs(((y_true[-1]-y_pred[-1])/y_true[-1])*100)))
```

```python
def euler_no_3(delta_t=0.1):
  f1_t_y=lambda y1,y2 : 1-y1+y2
  f2_t_y=lambda y1, y2 : 1+y1-y2
  t_0=0
  t_end=1
  # y_true_func=lambda t :2+3*np.exp(-t)
  # y_true=[y_0]
  # y_pred=[y_0]
  H = int((t_end-t_0)/delta_t)
  t=np.linspace(t_0, t_end, H+1)
  y1 = np.zeros(len(t))
  y2 = np.zeros(len(t))
  y1[0] = 0
  y2[0] = 0

  #looping
  for j in range(len(t)-1):
    y1[j+1]=y1[j]+f1_t_y(y1[j], y2[j])*delta_t
    y2[j+1]=y2[j]+f2_t_y(y1[j], y2[j])*delta_t
  print(y1[-1])
  print(y2[-1])
  # print('y_pred is\n{:0.5f}'.format(y_pred[-1]))
  # print('the error is {0:0.5f}%'.format(np.abs(((y_true[-1]-y_pred[-1])/y_true[-1])*100)))
```

```python
def euler_no_4(delta_t = 0.1):
  f_t_y_a=lambda y,a : 4-2*y-3*a
  t_0=0
  t_end=1
  # y_true_func=lambda t :2+3*np.exp(-t)
  # y_true=[y_0]
  # y_pred=[y_0]
  H = int((t_end-t_0)/delta_t)
  t=np.linspace(t_0, t_end, H+1)
  y = np.zeros(len(t))
  a = np.zeros(len(t))
  y[0] = 0
  a[0] = 0

  #looping
  for j in range(len(t)-1):
    y[j+1]=y[j]+a[j]*delta_t
    a[j+1]=a[j]+f_t_y_a(y[j],a[j])*delta_t
  print(y[-1])
  print(a[-1])
```
